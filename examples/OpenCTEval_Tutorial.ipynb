# OpenCTEval

OpenCTEval is an open-source framework for evaluating large language models (LLMs) on clinical trial tasks. It extends the capabilities of the LM-Evaluation-Harness to include multi-turn prompting strategies and specialized evaluation metrics tailored for the implemented clinical trial tasks.

In this notebook we will be going through a short tutorial on how things work. For full reference, please check out the original tutorials [lm-eval-overview](./original_examples/lm-eval-overview.ipynb).

## Run Model Evaluations

### Command Line Interface (CLI)

To run evaluations using OpenCTEval, we can use the command line interface (CLI) provided by the framework. Below is an example command to evaluate a model on a clinical trial task:

```bash
lm_eval --model hf \
    --model_args pretrained=Qwen/Qwen3-8B \
    --tasks MedNLI_0-shot \
    --device cuda:0 \
    --batch_size auto
```
